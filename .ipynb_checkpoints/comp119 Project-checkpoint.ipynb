```python
def train(X, T):
    # learned it from notebook, standardizing X 
    means = np.mean(X,axis=0)
    stds = np.std(X,axis=0)
    # X standard
    Xs = (X - means) / stds
    Xs = np.array(Xs).reshape((59,1))
    # add one column contains 1
    Xs1 = np.hstack((np.ones((Xs.shape[0],1)), Xs))                
    w = np.linalg.lstsq( np.dot(Xs1.T,Xs1), np.dot(Xs1.T, T) )[0]
    dict = {"means": means, "stds": stds, "w": w}
    return dict

def use(X, model):
    # learned it from notebook, always standardizing X
    Xs = (X - model["means"]) / model["stds"]
    Xs = np.array(Xs).reshape((2,1))
    # matrix multiplication
    return np.dot(np.hstack((np.ones((Xs.shape[0],1)), Xs)), model["w"])
```

```python
android_model = train(date[30:89], pop_android_pow[30:89])
android_prediction = use(date[89:], android_model)

# plot the results
plt.plot(date[30:],pop_android_pow[30:],'o',label="Real Android Popularity")
plt.plot(date[89:],android_prediction,'ro',label="Predicted Android Popularity")
plt.xlim(30, 95)
plt.legend(loc='upper right');
plt.xlabel('Date')
plt.ylabel('Popularity')
```
